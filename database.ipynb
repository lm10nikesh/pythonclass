```python
name=input("Enter the name: ")
physics= float(input("Enter the marks of physics: "))
chemistry= float(input("Enter the marks of chemistry: "))
maths= float(input("Enter the marks of maths: "))
english= float(input("Enter the marks of english: "))
nepali= float(input("Enter the marks of nepali: "))

total = physics+chemistry+maths+english+nepali
per = total/5

if per>=80:
    grade= 'A'
elif per>=60:
    grade= 'B'
elif per>= 45:
    grade= 'C'
else: 
    grade= 'F'

sql = f'''INSERT INTO student ( name, physics, chemistry, maths, english, nepali, total, per, grade) 
VALUES ( '{name}', {physics}, {chemistry}, {maths}, {english}, {nepali}, {total}, {per}, {grade});'''
db.execute(sql)
database.commit()  
```
